```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV

# Step 2: Upload dataset from hard drive
from google.colab import files
uploaded = files.upload()

# Replace with your actual filename
df = pd.read_csv("zambia_drought_forecasting_2015-2025.csv")

# Step 3: Inspect dataset
print(df.head())
print(df.columns)

# Expected columns: ['date','rainfall','soil_moisture','temperature','NDI']

# Step 4: Data Preprocessing
# Handle missing values (forward fill + linear interpolation)
for col in ['Rainfall_mm','SoilMoisture_pct','Temp_Max_C','NDVI']:
    df[col] = df[col].interpolate(method='linear').fillna(method='ffill')

# Normalize predictors (StandardScaler as per journal)
scaler = StandardScaler()
df[['Rainfall_mm','SoilMoisture_pct','Temp_Max_C']] = scaler.fit_transform(
    df[['Rainfall_mm','SoilMoisture_pct','Temp_Max_C']]
)

# Step 5: Feature Engineering
# Lag features
df['Rainfall_mm_lag7'] = df['Rainfall_mm'].shift(7)
df['SoilMoisture_pct_lag7'] = df['SoilMoisture_pct'].shift(7)

# Rolling averages
df['Rainfall_mm_roll14'] = df['Rainfall_mm'].rolling(window=14).mean()
df['Rainfall_mm_roll30'] = df['Rainfall_mm'].rolling(window=30).mean()

# Drop rows with NaN introduced by lag/rolling
df = df.dropna()

# Step 6: Feature Selection (basic correlation + ANOVA style)
# Pearson correlation with NDI
features = ['Rainfall_mm','SoilMoisture_pct','Temp_Max_C',
            'Rainfall_mm_lag7','SoilMoisture_pct_lag7',
            'Rainfall_mm_roll14','Rainfall_mm_roll30']

corr_results = {}
for f in features:
    r, p = pearsonr(df[f], df['NDVI'])
    corr_results[f] = (round(r,3), round(p,4))

print("Pearson Correlation Results:")
for f, vals in corr_results.items():
    print(f, "r =", vals[0], "p =", vals[1])

# Step 7: Model Development (Random Forest with hyperparameter tuning)
X = df[features]
y = df['NDVI']

# TimeSeriesSplit for temporal CV
tscv = TimeSeriesSplit(n_splits=5)

rf = RandomForestRegressor(random_state=42)

param_grid = {
    'n_estimators':[100,500],
    'max_depth':[10,20],
    'min_samples_split':[2,5]
}

grid_search = GridSearchCV(rf, param_grid, cv=tscv,
                           scoring='neg_root_mean_squared_error')
grid_search.fit(X,y)

print("Best Parameters:", grid_search.best_params_)
print("Best RMSE:", -grid_search.best_score_)
```

```python
import torch
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def gaussian_kernel(X, sigma=None):
    # X is (n, 1) tensor
    # Compute squared Euclidean distances
    sq_dist = torch.cdist(X, X, p=2)**2 # (n, n) matrix of squared Euclidean distances

    if sigma is None:
        # Median heuristic for sigma
        if X.numel() > 1:
            # Get unique distances from the upper triangle to avoid duplicates and zeros
            flat_dist = torch.triu(sq_dist, diagonal=1).flatten()
            non_zero_dist = flat_dist[flat_dist > 0]
            if non_zero_dist.numel() > 0:
                sigma = torch.sqrt(0.5 * torch.median(non_zero_dist))
            else:
                sigma = 1.0 # Default if all distances are zero or only one point
        else:
            sigma = 1.0 # Default for single point

    # Ensure sigma is not zero to prevent division by zero, use a small epsilon if needed
    if sigma == 0: sigma = 1e-6

    K = torch.exp(-sq_dist / (2 * sigma**2))
    return K

def hsic(x, y):
    x_tensor = torch.tensor(x.values, dtype=torch.float32).unsqueeze(1)
    y_tensor = torch.tensor(y.values, dtype=torch.float32).unsqueeze(1)

    n = x_tensor.size(0)
    H = torch.eye(n) - (1/n) * torch.ones(n, n)

    Kx = gaussian_kernel(x_tensor)
    Ky = gaussian_kernel(y_tensor)

    Kx_c = torch.mm(H, torch.mm(Kx, H))
    Ky_c = torch.mm(H, torch.mm(Ky, H))

    hsic_val = torch.trace(torch.mm(Kx_c, Ky_c)) / (n*n)
    return hsic_val.item()

# Define features if not already defined (assuming it's available from previous cells)
if 'features' not in locals() or not isinstance(features, list):
    features = ['Rainfall_mm', 'SoilMoisture_pct', 'Temp_Max_C']

hsic_values = []
for f in features:
    hsic_val = hsic(df[f], df['NDVI'])
    print(f"HSIC {f}-NDVI: {hsic_val:.3f}")
    hsic_values.append({'Feature': f, 'HSIC Value': hsic_val})

# Create a DataFrame for visualization
df_hsic = pd.DataFrame(hsic_values)

# Visualize HSIC values
plt.figure(figsize=(8, 5))
sns.barplot(x='Feature', y='HSIC Value', hue='Feature', data=df_hsic, palette='magma', legend=False)
plt.title('Hilbert-Schmidt Independence Criterion (HSIC) with NDVI (Gaussian Kernel)')
plt.xlabel('Features')
plt.ylabel('HSIC Value')
plt.ylim(0, df_hsic['HSIC Value'].max() * 1.1) # Set limits slightly above max for better visual
plt.axhline(0, color='black', linewidth=0.8)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
